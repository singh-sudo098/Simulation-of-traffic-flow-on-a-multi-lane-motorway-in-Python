```python
ID = 0

class P:
    def __init__(self, env, rec, 
                 inFront=None, 
                 t0=0, x0=0, dx0=0, ddx0=0, dddx0=0, 
                 t=[], v=[]):
        
        global ID
        self.id = ID
        ID += 1
        
        self.env = env
        self.rec = rec
        
        self.inFront = inFront
            
        self.x0 = x0
        self.dx0 = dx0
        self.ddx0 = ddx0
        self.dddx0 = dddx0
        
        self.t = t
        self.v = v
        self.t_target = []
        self.v_target = []
        
        # start process
        self.running = False
        self.processRef = None
        
        self.env.process(self.process(t0))             
    
    def updateOnly(self):
        t = self.env.now
        if t < self.t0:
            return False
        if self.running and t > self.t0:
            dt = t - self.t0
            ddx = self.ddx0 + self.dddx0*dt
            dx = self.dx0 + self.ddx0*dt + self.dddx0*dt*dt/2
            x = self.x0 + self.dx0*dt + self.ddx0*dt*dt/2 + self.dddx0*dt*dt*dt/6
            self.t0, self.x0, self.dx0, self.ddx0 = t, x, dx, ddx
        return True
                
    def process(self, t0):
        
        # delay start to the given time t-
        if t0 is not None and t0>self.env.now:
            yield self.env.timeout(t0-self.env.now)
        self.t0 = env.now
        self.running = True
        self.rec.startRecording(self)
        
        while self.running:
            self.updateOnly()
                
            if len(self.t_target)==0:
                self.t_target = self.t.copy()
                self.v_target = self.v.copy()
                
            if len(self.t_target)>0:
                v0 = self.dx0
                v1 = self.v_target[0]
                t = self.t_target[0]
                if isZero(v1-v0):
                    yield from self.wait(t)
                else:
                    yield from self.adjustVelocity(v1-v0, t)
                self.t_target = self.t_target[1:]
                self.v_target = self.v_target[1:]    
            else:
                yield from self.wait(1000)
                
        self.rec.stopRecording(self)
   
    # smoothly adjust velocity by Δv over the time Δt
    def adjustVelocity(self, Δv, Δt):
        self.updateOnly()
        minΔt = 0.1*Δt
        a = Δv/(Δt-minΔt)
        tt = Δt-2*minΔt
            
        self.dddx0 = (a-self.ddx0)/minΔt   
        yield self.env.timeout(minΔt)
        
        self.updateOnly()
        self.dddx0 = 0
        self.ddx0 = a
        yield self.env.timeout(tt)
        
        self.updateOnly()
        self.dddx0 = -a/minΔt
        yield self.env.timeout(minΔt)
        
        self.updateOnly()
        self.dddx0 = 0
        self.ddx0 = 0

    def wait(self, Δt):
        yield self.env.timeout(Δt)
```

```python
class SimpleRecorder:
    
    def __init__(self, env, startTime, stopTime, timeStep):
        
        global ID
        ID = 0
        
        self.env = env
        self.ps = []
        self.startTime = startTime
        self.stopTime = stopTime
        self.timeStep = timeStep
        self.data = pd.DataFrame(columns=['t', 'x', 'v', 'a', 'id', 'event'])

    # runs the simulation
    def run(self):
        self.env.process(self.process())
        self.env.run(self.stopTime+self.timeStep)
    
    def process(self):
        yield self.env.timeout(self.startTime-self.env.now)
        while self.env.now <= self.stopTime:
            self.record()     
            yield self.env.timeout(self.timeStep)
            
    def startRecording(self, p):
        self.ps.append(p)
        
    def stopRecording(self, p):
        self.ps.remove(p)
        
    def record(self):
        for p in self.ps:
            if p.updateOnly():
                ix = len(self.data)
                self.data.loc[ix]=[self.env.now, p.x0, p.dx0, p.ddx0, p.id, 'timer']

    def getData(self):
        return self.data.copy(deep=True)
            
    def plot(self, x, y, xmin=None, xmax=None, ymin=None, ymax=None):
        columns = ['t', 'x', 'v', 'a']
        labels = ['Time [s]', 'Position [m]', 'Velocity [m/s]', 'Acceleration [m/s²]']
        xindex = columns.index(x)
        yindex = columns.index(y)
        
        plt.figure(figsize=(6, 4), dpi=120)
        if xmin is not None and xmax is not None:
            plt.xlim((xmin, xmax))
        if ymin is not None and ymax is not None:
            plt.ylim((ymin, ymax))
        
        for p in self.ps:
            df = self.data[self.data.id==p.id]
            plt.plot(x, y, '', data=df)
            plt.xlabel(labels[xindex])
            plt.ylabel(labels[yindex])
            
        plt.grid(True)
```

```python
# To describe the slightly erratic speeding behaviour of human drivers we use the following approach:
# The driver is accellerating for a while, when suddenly the driver notice that it is getting too fast, the driver is decellerating. 
#The time intervals for these accellerating/decellerating patterns follows an expovariate distribution with average cycle time of 100s.
SLOW_CYCLE = 100
def randomIntervals(cycles):
    return [ max(0, random.normalvariate(SLOW_CYCLE, SLOW_CYCLE/3)) for i in range(cycles) ]
```

```python
randomIntervals(5)
```

```python
SPEED_VARIATION = 0.05
random.seed(13)
def randomSpeedVariation(vmax, cycles):
    return [ vmax + (-1)**i*abs(random.normalvariate(0, vmax*SPEED_VARIATION)) for i in range(cycles) ] 
```

```python
randomSpeedVariation(30,10)
```

```python
# Single point of varying speed
CYCLES = 6
VMAX = 120/3.6
random.seed(13)
env = simpy.Environment()
rec = SimpleRecorder(env, 0, 1000, 1)
times = randomIntervals(CYCLES)
speed = randomSpeedVariation(VMAX, CYCLES)
p = P(env, rec, dx0=speed[-1], t=times, v=speed)
rec.run()
```

```python
# Multiple points with varying speeds
CYCLES = 6
VMAX = 120/3.6
N = 5 # number of points
DT = 20 # time difference between start
random.seed(13)
env = simpy.Environment()
rec = SimpleRecorder(env, 0, 1000, 1)
p = None
for i in range(N):
    times = randomIntervals(CYCLES)
    speed = randomSpeedVariation(VMAX, CYCLES)
    p = P(env, rec, inFront=p, t0=i*DT, dx0=VMAX, t=times, v=speed)
rec.run()
```
